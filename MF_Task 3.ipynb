```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
import numpy as np
```

```python
house_df=pd.read_csv("householdtask3.csv")
house_df.head()
```

```python
plt.style.use('seaborn-v0_8');
```

```python
fig,ax0=plt.subplots()
ax0.bar(house_df["year"],house_df["income"])
ax0.set(title="Comparision of increase of Income over the year 2008-20",xlabel="Year",ylabel="Total Income");
```

```python
fig,ax1=plt.subplots()
aggregated_df = house_df.groupby('year').agg({'income': 'sum', 'expenditure': 'sum'}).reset_index()
ax1.plot(aggregated_df["year"], aggregated_df["expenditure"], label='Cumulative Expenditure')
ax1.plot(aggregated_df["year"],aggregated_df["income"],label="Commulative Income")
ax1.set(title="Cumulative Income and Expenditure Over Years",xlabel="year",ylabel="Amount ($)")
plt.legend();
```

```python
years = house_df['year'].unique()

for year in years:
    year_data = house_df[house_df['year'] == year]
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
    ax1.hist(year_data['own_prop'], bins=10, color='blue', edgecolor='black')
    ax1.set_title(f'Histogram of own_prop for {year}')
    ax1.set_xlabel('own_prop')
    ax1.set_ylabel('Frequency')

    ax2.hist(year_data['own_wm_prop'], bins=10, color='green', edgecolor='black')
    ax2.set_title(f'Histogram of own_wm_prop for {year}')
    ax2.set_xlabel('own_wm_prop')
    ax2.set_ylabel('Frequency')
```
